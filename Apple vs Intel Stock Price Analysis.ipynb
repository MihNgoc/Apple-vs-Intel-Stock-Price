```python
# import library
import pandas as pd
import numpy as np

# import library for visualisation
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
stock_price = pd.read_csv('./prices.csv')
```

```python
stock_price.head()
```

## Apple

```python
apple = stock_price[stock_price['symbol']=='AAPL']
```

```python
apple
```

```python
plt.figure(figsize=(18, 8))
apple['date'] = pd.to_datetime(apple['date'])

plt.plot(apple['date'], apple['close'])
plt.plot(apple['date'], apple['open'])
plt.xticks(rotation=30)
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Open and close prices of Apple stocks (2010-2017)')
plt.legend(['Open', 'Close'])
plt.show()
```

- We notice a sharply drop in the prices of shares in the middle of 2014 due to:
    - The stock split
    - Slower iPhone sales
    - Concerns about Apple's future growth

- After a fell, the stock price remain stable at that price until 2017.


```python
plt.figure(figsize=(18, 8))
apple['date'] = pd.to_datetime(apple['date'])

plt.plot(apple['date'], apple['volume'])

plt.xticks(rotation=30)
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Volume of stock traded (2010-2017)')
plt.show()
```

- At the beginning of every year, there was a sharply increase in the volume of stock traded. Followed by a downward trend in the trading transaction throughout the whole year. 
- We can clearly see a reduction in the volume of stocks traded after 2014

## Intel

```python
intel = stock_price[stock_price['symbol']=='INTC']
```

```python
intel.head()
```

```python
plt.figure(figsize=(18, 8))
intel['date'] = pd.to_datetime(intel['date'])

plt.plot(intel['date'], intel['close'], 'b')
plt.plot(intel['date'], intel['open'], 'r')
plt.xticks(rotation=30)
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Open and close prices of Intel stocks (2010-2017)')
plt.legend(['Open', 'Close'])
plt.show()
```

- There was a steady upward trend for the stock price of Intel from 2013 till 2017
- A clearly steep rise around May 2014 due to:
    - Strong earnings
    - Positive outlook
    - M&A activity

```python
plt.figure(figsize=(18, 8))
intel['date'] = pd.to_datetime(intel['date'])

plt.plot(intel['date'], intel['volume'])

plt.xticks(rotation=30)
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Volume of stock traded (2010-2017)')
plt.show()
```

- At the beginning of every year, there was a sharply increase in the volume of stock traded. Followed by a downward trend in the trading transaction throughout the whole year. 
- We can clearly see a reduction in the volume of stocks traded after 2014

## Conclusion

- In 2014, while there was a sharply drop in Apple's stock price, the stock price of Intel started a steady increase till 2017
- From 2010 till the end of 2013, Intel and Apple both have a large voulme of strok traded
- However, from 2014, there was a reduction in the volume of stock traded due to: 
    - Low interest rates make it less attractive to invest in stocks
    - Rising stock prices (especially, Intel)
    - Investor uncertainty
    - Increased regulation of the financial industry can make it more difficult and costly for investors to trade
